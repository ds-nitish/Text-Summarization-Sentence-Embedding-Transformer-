# Text Summarization

Web scrapping data from wikipedia to do summarization

```python
import requests
from bs4 import BeautifulSoup
```

```python
url ='https://en.wikipedia.org/wiki/Transformer_(machine_learning_model)'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content  , 'html.parser')
```

```python
print(soup.prettify())
```

```python
tags = soup.find_all('p')
text = [i.text for i in tags]
```

```python
# Cleaning Data
import re
val =[]
for i in text:
  pattern = r"[^a-zA-Z0-9\s]"
  data = re.sub(pattern,'',i)
  data = re.sub('[\n]','',i)
  data = ''.join(data)
  val.append(data)
```

```python
val
```

```python
val = val[:15]
article = ''.join(val)
```

```python
# Article to summarize
article
```

```python
print(article)   ## This our text which we want to summarize
```

```python
# creating a model which work on bert to get sentence embeddings
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
# Creating list of sentences
sentences = article.split('.')
```

```python
len(sentences)
```

```python
# Getting array of all sentence in list
# we can also do embedding for paragraph.we have to do embedding of all sentences in paragraph and after that aggrigate it together to get embedding of paragraphs-----similar to avg word2vec
sentence_embeddings = model.encode(sentences)
```
